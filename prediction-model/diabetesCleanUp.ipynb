```python
# DIABETE4
# going to make this ordinal. 0 is for no diabetes or only during pregnancy, 1 is for pre-diabetes or borderline diabetes, 2 is for yes diabetes
# Remove all 7 (dont knows)
# Remove all 9 (refused)
# Remove NaN values in the DIABETE4 column
brfss_df_selected_no_nan = brfss_df_columned.dropna(subset=['DIABETE4'])

# Replace DIABETE4 values with ordinal categories
brfss_df_selected_no_nan.loc[:, 'DIABETE4'] = brfss_df_selected_no_nan['DIABETE4'].replace({2: 0, 3: 0, 4: 1, 1: 2})

# Remove rows where DIABETE4 is 7 (Don't know) or 9 (Refused)
brfss_df_labeled = brfss_df_selected_no_nan[~brfss_df_selected_no_nan['DIABETE4'].isin([7, 9])]

# Verify the value counts
print(brfss_df_labeled['DIABETE4'].value_counts())
print(brfss_df_labeled.shape)

# Save the cleaned DataFrame
brfss_df_selected = brfss_df_labeled
```
